```python
p1 = pd.read_csv('perc1.csv', encoding = "utf8")
p2 = pd.read_csv('perc2.csv', encoding = "utf8")
```

```python
values = p1.perc
clrs = ['#B2DD2C' if ((x < 0.2) & (x>-0.4)) else '#472777' for x in values]

fig = sb.barplot(x = 'countryS', y='perc',data=p1, palette=clrs,order= ['GB','FR','IT','ES','DE','BE','PT','NL'])
sb.despine(left=True)
fig.set(xlabel='Percentage sales price compared to Germany', ylabel='Seller country')
```

```python
values2 = p2.perc
clrs2 = ['#472777' if (x>0.2) else '#B2DD2C' for x in values2]

fig = sb.barplot(x='countryB', y= 'perc',data=p2,palette=clrs2,order= ['GB','FR','IT','ES','DE','BE','PT','NL'])
sb.despine(left=True)
fig.set(xlabel='Percentage purchase price compared to Germany', ylabel='Buyer country')
```
