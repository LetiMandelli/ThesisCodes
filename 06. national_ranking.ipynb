# Modify columns: 1516 1617 1718
I always consider the mean of the results obtained in 15 days of championship. So we have to modify the columns relative to the whole year, according to the number of days in each league. 38 for spa, eng, ita, fra. 34 for ger, ola, port and 30 for belgium

```python
# 0 al posto di na: se c'è 0 sono in serie B
d_tf2['1516'] = d_tf2['1516'].fillna(0)
d_tf2['1617'] = d_tf2['1617'].fillna(0)
d_tf2['1718'] = d_tf2['1718'].fillna(0)

def meanOn15Days1516(df):
    if df['country_selling'] == 'GB' or df['country_selling'] == 'IT' or df['country_selling'] == 'FR' or df['country_selling'] == 'ES':
        return round(df['1516']*15/38)
    elif df['country_selling'] == 'NL' or df['country_selling'] == 'PR' or df['country_selling'] == 'DE':
        return round(df['1516']*15/34)
    elif df['country_selling'] == 'BE':
        return round(df['1516']*15/30)
    else:
        return np.nan

d_tf2['1516'] = d_tf2.apply(meanOn15Days1516, axis = 1)

def meanOn15Days1617(df):
    if df['country_selling'] == 'GB' or df['country_selling'] == 'IT' or df['country_selling'] == 'FR' or df['country_selling'] == 'ES':
        return round(df['1617']*15/38)
    elif df['country_selling'] == 'NL' or df['country_selling'] == 'PR' or df['country_selling'] == 'DE':
        return round(df['1617']*15/34)
    elif df['country_selling'] == 'BE':
        return round(df['1617']*15/30)
    else:
        return np.nan
    
d_tf2['1617'] = d_tf2.apply(meanOn15Days1617, axis = 1)

def meanOn15Days1718(df):
    if df['country_selling'] == 'GB' or df['country_selling'] == 'IT' or df['country_selling'] == 'FR' or df['country_selling'] == 'ES':
        return round(df['1718']*15/38)
    elif df['country_selling'] == 'NL' or df['country_selling'] == 'PR' or df['country_selling'] == 'DE':
        return round(df['1718']*15/34)
    elif df['country_selling'] == 'BE':
        return round(df['1718']*15/30)
    else:
        return np.nan

d_tf2['1718'] = d_tf2.apply(meanOn15Days1718, axis = 1)
```

```python
d_tf2['nationalPerf_team_from'] = np.nan

def perfYearBefore(df):
    if df['year'] == 2016 and df['wintersummer']=='s':
        return df['1516']
    elif df['year'] == 2017 and df['wintersummer']=='s':
        return df['1617']
    elif df['year'] == 2018 and df['wintersummer']=='s':
        return df['1718']
    elif df['year'] == 2017 and df['wintersummer']=='w':
        return df['1617_half']
    elif df['year'] == 2018 and df['wintersummer']=='w':
        return df['1718_half']
    elif df['year'] == 2019 and df['wintersummer']=='w':
        return df['1819_half']
    else:
        return np.nan

d_tf2['nationalPerf_team_from'] = d_tf2.apply(perfYearBefore, axis = 1)
```

```python
# drop columns with the rankings of all years
d_tf2 = d_tf2.drop(['1516','1617_half','1617','1718_half','1718','1819_half'], axis=1)
```

# Dummy creation: 0 if it was not in the 1st division when the player was sold

```python
d_tf2['nationalPerf_team_from'] = d_tf2.nationalPerf_team_from.fillna(0)

d_tf2['dummyAB'] = np.nan

def dummyAB(df):
    if df['nationalPerf_team_from'] == 0:
        return 0
    else:
        return 1
    
# 0 = not A, 1 = A
d_tf2['dummyAB'] = d_tf2.apply(dummyAB, axis = 1)

d_tf2.head()
```
